```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import linear_model
from sklearn.model_selection import train_test_split

df = pd.read_csv("gold.csv")
data = df.dropna()
fig, axs = plt.subplots(1,3, figsize = (18,9))
data.plot(kind='scatter', x = 'USD (AM)', y = 'USD (PM)', ax = axs[0])
data.plot(kind='scatter', x = 'GBP (AM)', y = 'GBP (PM)', ax = axs[1])
data.plot(kind='scatter', x = 'EURO (AM)', y = 'EURO (PM)', ax = axs[2])
```

```python
x = np.array(data['USD (AM)'])
y = np.array(data['USD (PM)'])
x_mean, y_mean = np.mean(x), np.mean(y)
denom = numer = 0
for i in range(len(x)):
    numer += (x[i] - x_mean)*(y[i] - y_mean)
    denom += (x[i] - x_mean)**2
m = numer/denom
c = y_mean - (m*x_mean)
```

```python
xMax = np.max(x)
xMin = np.min(x)

X = np.linspace(xMin, xMax, len(x))
Y = c + m*X
plt.plot(X, Y, color = "red", label = "regression line")
plt.scatter(X, Y, color = "blue", label = "scatter plot")

plt.show
```

```python
ss_t = ss_r = 0
for i in range(len(x)):
    y_pred = c + m*x[i]
    ss_t +=(y[i]-y_mean)**2
    ss_r +=(y[i]-y_pred)**2
r2 = 1 - (ss_r/ss_t)
r2
```
